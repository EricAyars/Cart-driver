```python
import matplotlib.pyplot as plt
from numpy import loadtxt
```

```python
f, C1A, C1P, C2A, C2P = loadtxt("Sweep.txt", unpack=True, skiprows=7)
```

```python
fig,ax1=plt.subplots()
ax1.plot(f, C1A, 'g-', label="Cart 1 amplitude")
ax1.plot(f, C2A, 'b-', label="Cart 2 amplitude")
ax1.set_xlabel('Frequency (Hz)')
ax1.set_ylabel('Amplitude (cm)')

ax2 = ax1.twinx()
ax2.plot(f, C1P, 'g--', label="Cart 1 phase")
ax2.plot(f, C2P, 'b--', label="Cart 2 phase")
ax2.set_ylabel("Angle (radians)")

plt.title("Frequency response of 2-cart/spring system")
```
